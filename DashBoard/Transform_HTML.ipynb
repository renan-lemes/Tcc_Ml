```python
import pandas as pd
import numpy as np
from datetime import datetime, date
from IPython.display import HTML
import pandas_datareader as web
```

```python
stock = 'SPY'
today = datetime.now()
one_year = today.replace(year=today.year-1)
```

```python
df = web.DataReader(stock,'yahoo', one_year, today)
df = pd.DataFrame(df)
```

```python
df = df.reset_index()
df
```

```python
results = df.to_html()
```
